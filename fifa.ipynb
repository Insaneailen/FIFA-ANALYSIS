```python
fifa_df['Preferred Foot'] = fifa_df['Preferred Foot'].astype(str)
```

```python
loopStartTime = time.time()
leftCount = 0
rightCount = 0
nanCount = 0
for index, row in fifa_df.iterrows():
    if row['Preferred Foot'] == 'Left' :
        leftCount += 1
    elif row['Preferred Foot'] == 'Right':
        rightCount += 1
    elif row['Preferred Foot'] == 'nan':
        nanCount += 1
loopEndTime = time.time()     
print(f'There are: {leftCount} left foot players \
      , and {rightCount} right foot players, we also have {nanCount} unknown values\
        , and execution takes {loopEndTime - loopStartTime}')
```
